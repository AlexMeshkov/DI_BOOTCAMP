### Part 2

```python
df = load_boston
new_properties = df[df['age'] <= 35]
old_properties = df[(df['age'] > 35) & (df['age'] <= 70)]
very_old_properties = df[df['age'] > 70]
```

```python
figure, axes = plt.subplots(3,1)
figure.tight_layout
axes[0].set(xlabel='median value', ylabel='age')
axes[0].scatter(new_properties['medv'], new_properties['age'])
axes[1].set(xlabel='median value', ylabel='age')
axes[1].scatter(old_properties['medv'], old_properties['age'])
axes[2].set(xlabel='median value', ylabel='age')
axes[2].scatter(very_old_properties['medv'], very_old_properties['age'])
plt.show()
```

### Part 3.

```python
load_boston['CRIM_BIN'] = pd.cut(load_boston['crim'], bins=[0, 1, 5, 10, 20, max(load_boston['crim'])], labels=['0-1', '1-5', '5-10', '10-20', '20+'])
load_boston
```

```python
avg_MEDV_per_CRIM_BIN = load_boston.groupby('CRIM_BIN')['medv'].mean()
avg_MEDV_per_CRIM_BIN
```

```python
figure, axes = plt.subplots()
avg_MEDV_per_CRIM_BIN.plot(kind='line', ax=axes, xlabel='average crime', ylabel='median value')
plt.show()
```
