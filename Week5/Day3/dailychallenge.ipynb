### Task1

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### 1.1

```python
figure, axes = plt.subplots()
x = np.linspace (0, 21 )
y = x**2
axes.plot(x,y)
axes.set(title = 'Plot of x vs y(=x^2)', xlabel = "X", ylabel = "Y")
figure.savefig('Plot of x vs y.png')
```

### Task2

```python
chipo = pd.read_csv('chipo.tsv', sep = '\t')
```

```python
chipo
```

```python
chipo['item_price'] = chipo['item_price'].apply(lambda row: float(row[1:]))
figure, axes = plt.subplots()
x = chipo['item_price']
y = chipo['quantity']
axes.plot(x,y, marker = 'd')
axes.set (xlabel = 'Item price', ylabel = 'Quantity ordered', title = 'Relationship between Item Price and Quantity Ordered')
figure.savefig ('Relationship between Item Price and Quantity Ordered.png')
```
