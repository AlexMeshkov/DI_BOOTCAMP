### Exercise 7

```python
x = np.linspace(-10, 10, 2000)
y1 = np.sin(x )
y2 = np.cos(x)
figure, axes = plt.subplots()
axes.set(title = "y = sin(x)")
axes.plot(x,y1, 'tab:blue')
```

```python
figure, axes = plt.subplots()
axes.set(title = "y = cos(x)")
axes.plot(x,y2, color ='green', marker = 'd')
```

### Exercise 8

```python
figure, axes_set = plt.subplots(2)

axes_set[0].plot(x, y1, color='tab:blue')
axes_set[1].scatter(x, y2, color='green', marker='d')

plt.tight_layout()
```

### Exercise 9

```python
figure, axes = plt.subplots(2)

axes[0].plot(x, y1, color='tab:blue', label = 'y = sin(x)')
axes[1].scatter(x, y2, color='green', marker='d', label = 'y = cos(x)')
axes[0].legend()
axes[1].legend(loc = 'center right', shadow = True)
plt.tight_layout()
```
